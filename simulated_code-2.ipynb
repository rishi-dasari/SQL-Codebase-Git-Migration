```python
import os
import random
import shutil
import subprocess

# ==========================================
# Phase 1: Setup & Mock Data Generation
# ==========================================
LEGACY_DIR = '/content/legacy_sql_share'
GIT_DIR = '/content/sql_codebase_repo'
FOLDERS = ['Tables', 'Views', 'StoredProcedures', 'Functions']

print("🧹 Cleaning up old simulation runs...")
shutil.rmtree(LEGACY_DIR, ignore_errors=True)
shutil.rmtree(GIT_DIR, ignore_errors=True)

print("📁 Creating legacy shared folders...")
os.makedirs(LEGACY_DIR, exist_ok=True)
for f in FOLDERS:
    os.makedirs(os.path.join(LEGACY_DIR, f), exist_ok=True)

print("📄 Generating 4,000 mock SQL objects...")
for i in range(1, 4001):
    obj_type = random.choice(FOLDERS)
    filename = f"{obj_type[:-1]}_{i}.sql"
    filepath = os.path.join(LEGACY_DIR, obj_type, filename)
    with open(filepath, 'w') as f:
        f.write(f"-- Legacy SQL Object: {filename}\n")
        f.write(f"-- Created by: Unknown User (Pre-Git Era)\n")
        f.write(f"-- Last Modified: 2018-05-12\n\n")
        f.write(f"CREATE {obj_type[:-1].upper()} dbo.{filename.replace('.sql', '')} AS\n")
        f.write("SELECT * FROM dbo.DummyTable;\nGO\n")

print(f"✅ Successfully generated 4,000 legacy SQL files in {LEGACY_DIR}\n")

# ==========================================
# Phase 2: Git Initialization & Migration
# ==========================================
print("🚀 Initializing new Git Repository...")
os.makedirs(GIT_DIR, exist_ok=True)

def run_git(args, cwd=GIT_DIR):
    result = subprocess.run(["git"] + args, cwd=cwd, capture_output=True, text=True)
    if result.returncode != 0:
        print(f"Git Error: {result.stderr}")
    return result.stdout.strip()

# Setup git config (Replace with your info in Colab if needed)
run_git(["config", "--global", "user.name", "Data Engineer"])
run_git(["config", "--global", "user.email", "engineer@portfolio.local"])

# Initialize repo and create main branch
run_git(["init"])
run_git(["checkout", "-b", "main"])

print("📦 Migrating files from shared drive to Git structure...")
# In a real scenario, you might restructure files here (e.g., schema/type folders)
for item in os.listdir(LEGACY_DIR):
    s = os.path.join(LEGACY_DIR, item)
    d = os.path.join(GIT_DIR, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)

# Initial Commit
run_git(["add", "."])
run_git(["commit", "-m", "feat: Initial migration of 4,000 SQL objects from legacy shared drive"])

# Create Baseline Release Tag
run_git(["tag", "-a", "v1.0.0", "-m", "Baseline Production Release"])
print("✅ Committed 4,000 files and tagged release v1.0.0\n")

# ==========================================
# Phase 3: Simulating the New Workflow
# (Branching, PRs, and Hotfixes)
# ==========================================
print("🔀 Simulating a developer creating a feature branch...")
run_git(["checkout", "-b", "feature/update-customer-view"])

# Modify an existing view
view_path = os.path.join(GIT_DIR, 'Views')
view_files = os.listdir(view_path)
if view_files:
    target_view = os.path.join(view_path, view_files[0])
    with open(target_view, 'a') as f:
        f.write("\n-- Updated by Feature Branch: Added Customer Segment Logic\n")

    run_git(["add", target_view])
    run_git(["commit", "-m", "refactor: updated customer view logic"])
    print(f"✅ Modified {view_files[0]} on feature branch.")

print("🔀 Simulating PR merge into main...")
run_git(["checkout", "main"])
run_git(["merge", "--no-ff", "feature/update-customer-view", "-m", "Merge pull request #1 from feature/update-customer-view"])

# Tag new release
run_git(["tag", "-a", "v1.1.0", "-m", "Release v1.1.0 - Customer View Updates"])
print("✅ Merged PR and tagged release v1.1.0\n")

# ==========================================
# Phase 4: Simulating a Rollback
# ==========================================
print("🚨 Simulating a bad deployment and rollback...")
print(f"Current Head: {run_git(['log', '-1', '--oneline'])}")
print("Rolling back to v1.0.0...")
run_git(["checkout", "v1.0.0"])
print(f"Rollback Head: {run_git(['log', '-1', '--oneline'])}")

print("\n🎉 SIMULATION COMPLETE! Repository is ready in /content/sql_codebase_repo")
```
